# setup environment

```r
library(tidyverse)
```

```r
library(corrplot)
```

```r
datadir <- "/data/DataFest2017"
```

# import data

```r
dat_xped <- read_delim(file.path(datadir, "data_subset_prob20.txt"), delim = "\t")
```

```r
head(dat_xped)
```

```r
dat_dest <- read_delim(file.path(datadir, "dest.txt"), delim = "\t")
```

```r
head(dat_dest)
```

# compare the location

```r
dat <- dat_dest %>% select(contains("activity"))
head(dat)
```

```r
mat_cor <- cor(t(dat), method="spearman")
```

```r
head(mat_cor)
```

```r
dim(mat_cor)
```

```r
corrplot(mat_cor[1:100, 1:100], is.corr = TRUE)
```
